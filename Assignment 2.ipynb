## Descriptive Analytics and Data Preprocessing on Sales & Discounts Dataset

### Descriptive Analytics for Numerical Columns

##### ‚óè	Objective: To compute and analyze basic statistical measures for numerical columns in the dataset.
##### Steps:
	 Load the dataset into a data analysis tool or programming environment (e.g., Python with pandas library).
	 Identify numerical columns in the dataset.
	 Calculate the mean, median, mode, and standard deviation for these columns.
	 Provide a brief interpretation of these statistics.


```python
import pandas as pd
df = pd.read_csv(r'C:\Users\DELL\Documents\sales_data_with_discounts.csv')
```

```python
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
print("Numerical Columns:", numerical_columns)
```

```python
for col in numerical_columns:
    mean = df[col].mean()
    median = df[col].median()
    mode = df[col].mode()[0]
    std_dev = df[col].std()
    print(f'{col} - Mean: {mean}, Median: {median}, Mode: {mode}, Std Dev: {std_dev}')
```

```python
#### Interpretation
# A higher standard deviation indicate more variability in the data
```

## Data Vizualization

##### Objective: To visualize the distribution and relationship of numerical and categorical variables in the dataset.

#### Histogram
##### Plot histograms for each numerical column. Analyze the distribution (e.g., skewness, presence of outliers) and provide inferences

```python
# Histograms for each numerical column to visualize the distribution and check for skewness
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for col in numerical_columns:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], bins=20, kde= True)
    plt.title(f'Histogram of {col}')
    plt.show()
```

```python
# In the histogram of Volume, Average price, Total sales value, Discount amount and Net sales value there is positive skewness.
# In the histogram of Discount rate there is negative skewness.
# This chart also has outliers in all the columns
```

#### Boxplots:
##### 	Create boxplots for numerical variables to identify outliers and the interquartile range. Discuss any findings, such as extreme values or unusual distributions.

```python
# Boxplots to identify outliers and the interquatile range (IQR).
```

```python
for col in numerical_columns:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
#### Interpret Findings:
# Outliers are values that fall outside the 1.5*IQR range. In boxplot of Volume outlier should be 30.0 or after 30 dot. In Avg Price the 60000 is outlier. In Total sales 200000 is outlier. In DIscount rate around 4.23 something should be outlier. In Discount amount 25000 is outlier. In boxplot of Net Sales value 175000 is outlier.
# Extreme values or asymmetry in the boxplot may indicate skewed data
# IQR: In Boxplot of Volume IQR is 4.5, In Avg price IQR is must be 100, In Total Sales value IQR is 0.4345, In Discount rate IQR is 16.5. etc
```

#### Bar Chart Analysis for Categorical Column:
	Identify categorical columns in the dataset.
	Create bar charts to visualize the frequency or count of each category.
	Analyze the distribution of categories and provide insights.


```python
# Identifying Categorical Columns:
categorical_columns = df.select_dtypes(include=['object']).columns
print("Categorical Columns:", categorical_columns)
```

## Standardization of Numerical Variables

#### Objective: To scale numerical variables for uniformity, improving the dataset‚Äôs suitability for analytical models.
#### Steps:
	Explain the concept of standardization (z-score normalization).
	Standardize the numerical columns using the formula: z=x-mu/sigma
	
	Show before and after comparisons of the data distributions


```python
### Concept:
# Standardization transform numerical variables to have a mean of 0 and a standard deviation of 1.
```

```python
#### Apply Standardization:

from sklearn.preprocessing import StandardScaler

scaler= StandardScaler()
df_scaled = df.copy()
df_scaled[numerical_columns] = scaler.fit_transform(df[numerical_columns])
```

```python
### Before and After Comparisons: we can compare the distribution before and after scaling using histograms.

for col in numerical_columns:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], bins=20, kde=True, label='Standardized', color='red')
    plt.title(f'Original vs Standardized {col}')
    plt.legend()
    plt.show()
```

### Conversion of Categorical Data into Dummy Variables

```python
#### CONCEPT: Categorical data must be converted into dummy variables (one-hot encoding) for machine learning models. 
```

```python
#### Apply One-Hot Encoding:
df_dummies = pd.get_dummies(df, columns=categorical_columns, drop_first=True)
print(df_dummies.head())
```

Concept of One-Hot Encoding

```python
# Categorical variables are transformed into binary columns,with each unique category having its own columns.
# One-hot encoding ensures that no artificial order is imposed on categorical data,which could mislead certain models.
### IMPORTANCE:
#It allows machine learning algorithms to use categorical data correctly without introducing ordinall relationships.
# Some models, like decision trees, can handle categorical data natively,but most models (e.g., linear models)require this preprocessing step.
```
