###  Data Cleaning and Preparation

```python
# First we need to load the dataset
import pandas as pd
df = pd.read_csv(r'C:\Users\DELL\Documents\Cardiotocographic.csv')
df.head(5)
```

##### Handle missing value

```python
# Checking for missing value
df.isnull().sum()   #to handle missing value
```

```python
df.fillna(df.mean(), inplace=True)   # Imputation with mean
df.mean()
```

```python
df.fillna(df.median())     # Imputation with meadian
df.median()
```

```python
df.fillna(df.dropna())
df.dropna()
```

#### Detecting and Treating Outliers:


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Boxplot to visualize outliers
plt.figure(figsize=(12,6))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
#sns.boxplot(data=df): Draws boxplots to visualize outliers.
# There are number of outliers for instance in LB, ASTV, ALTV, MLTV, Width etc
```

##### Outlier Treatment using the IQR Method

```python
#####Interquartile Range (IQR) Approach:
#Step 1: Compute the first (Q1) and third (Q3) quartiles.
#Step 2: Calculate the IQR = Q3 - Q1.
#Step 3: Define lower and upper bounds as:
       #Lower bound = Q1 - 1.5 × IQR
       #Upper bound = Q3 + 1.5 × IQR
#Step 4: Replace outliers with boundary values.
```

```python
## Calculate IQR
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
```

```python
Q1
```

```python
Q3
```

```python
IQR
```

```python
# Defining bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
lower_bound
```

```python
upper_bound
```

```python
# Cap values within the boundaries
df = df.clip(lower=lower_bound, upper=upper_bound, axis=1)
df
```

```python
#The IQR method ensures extreme values don’t distort the analysis.
```
