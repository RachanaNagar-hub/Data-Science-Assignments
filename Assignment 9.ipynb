### Task 2: Dimensionality Reduction with PCA

##### 1. Standardize the Features:

```python
#PCA requires the features to be standardized so that they all have a mean of 0 and standard deviation of 1.

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()      # Standardizing the dataset
data_scaled = scaler.fit_transform(df)
data_scaled
```

##### 2. Implement PCA:

```python
#PCA reduces the dimensionality by transforming the features into new principal components that capture the most variance in the data.

from sklearn.decomposition import PCA

pca = PCA()
pca.fit(data_scaled)
explained_variance = pca.explained_variance_ratio_
explained_variance
```

```python
# Scree Plot
plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.title('Scree Plot')
plt.show()
```

##### 3.Transform Dataset into Principal Components:

```python
# Cumulative explained variance
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)
chosen_number_of_components = np.argmax(cumulative_variance >= 0.90) + 1  # for 90% variance

# Choose number of components, e.g., enough to explain 90% of variance
pca = PCA(n_components=chosen_number_of_components)
data_pca = pca.fit_transform(data_scaled)
```

### Task 3: Clustering with Original Data

##### 1. Apply a Clustering Algorithm:

```python
# Using K-Means clustering and determine the optimal k using the elbow method or silhouette score.

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

optimal_k = 4 
kmeans = KMeans(n_clusters=optimal_k)
kmeans.fit(df)
labels = kmeans.labels_
```

##### 2.Visualize Clustering Results:

```python
# Using pair plots or t-SNE for high-dimensional data to visualize clusters.
```

```python
# Example visualization for 2 features
sns.scatterplot(x=df['Flavanoids'], y=df['Malic'], hue=labels)
plt.show()
```

##### 3.Evaluate Clustering Performance:

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score

silhouette = silhouette_score(df, labels)      ## Calculate silhouette or Davies-Bouldin scores
db_index = davies_bouldin_score(df, labels)
db_index
```

### Task 4: Clustering with PCA Data

```python
# Applying the clustering algorithm to the PCA-transformed data.
kmeans_pca = KMeans(n_clusters=optimal_k)
kmeans_pca.fit(data_pca)
labels_pca = kmeans_pca.labels_
kmeans_pca
```

```python
# Visualize the clusters in the principal component space.

# Plot using first two principal components
plt.scatter(data_pca[:, 0], data_pca[:, 1], c=labels_pca)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Clustering on PCA Data')
plt.show()
```
